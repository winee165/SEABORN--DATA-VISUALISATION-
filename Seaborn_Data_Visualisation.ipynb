# **SEABORN PYTHON LIBRARY**

Importing Seaborn and other necessary libraries

```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
```

**We will perforn Data Visualisation on "TIPS Dataset".**


Tips Dataset is openly available. So, we just need to load the dataset.

## **SCATTERPLOT**

```python
sns.set(style="darkgrid")
tips = sns.load_dataset("tips")   #Load dataset
print(tips)                       #print dataset

# Seaborn's default plot is "scatterplot".
sns.relplot(x="total_bill", y="tip",label="bill vs tip",data=tips);
```

Hue : produce data points with different colors. To group to multiple data variable and show the dependency of the passed data values are to be plotted

```python
sns.relplot(x="total_bill", y="tip",hue="smoker", data=tips);
```

Size : Adjusts the size of data points for better visual.

```python
sns.relplot(x="total_bill", y="tip",hue="smoker",size="size", data=tips);
```

# **LINEPLOT**

```python
sns.relplot(x="total_bill",y="tip",kind="line",color="hotpink",data=tips)
```

Sort: Lineplot assumes that data points are in sorted order.

```python
sns.relplot(x="total_bill",y="tip",kind="line",sort=False,data=tips)
```

Col/Row : Categorical variables that will determine the faceting of the grid.

```python
sns.relplot(x="total_bill", y="tip",hue="smoker",size="size",col="sex", data=tips);
```

## CATPLOT 

```python
sns.catplot(x="total_bill",y="sex",data=tips)
```

# SWARMPLOT

```python
sns.catplot(x="smoker",y="total_bill",kind="swarm",hue="sex",data=tips)
```

# BOXPLOT

```python
sns.catplot(x="total_bill",y="day",kind="box",data=tips)
```

```python
sns.catplot(x="total_bill",y="day",kind="box",hue="smoker",data=tips)
```

# BARPLOT

```python
sns.catplot(x="sex",y="tip",kind="bar",data=tips)
```

```python
sns.catplot(x="sex",y="tip",kind="bar",hue="smoker",data=tips)
```

COUNT PLOT: Special type of Bar Graph that counts the occurrence of a particular type of data.

```python
sns.catplot(x="day",kind='count',data=tips)
```

# POINTPLOT

```python
sns.catplot(x="day",y="tip",kind="point",hue="sex",data=tips,markers="o")
```

Merging several functionalities/attributes in a single graph.

```python
sns.catplot(x="tip",y="time",kind="bar",col="sex",hue="smoker",data=tips)
```
